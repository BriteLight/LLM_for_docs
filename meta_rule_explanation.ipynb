```python
prompts = {
        "PREFIX": """You are an AI Agent which help in explaing the meta rule with the following tools at your disposal:
        VariableDict: Retrieve the mapping i.e dictionary of variable descriptions required for machine learning model decision from order_details dictionary. Input shall be a list of variables""",

        "SUFFIX": """When you answer, follow these steps meticulously:

        1. Identify the  list of variables required for explaining the meta rule \n
        2. Use the VariableDict tool to get the list of descriptions required for explaining the meta rule \n
        
    Using the following example try to explain the meta_rule
    Example:

    meta_rule: 
            
    var A = (_dv_is_tmx_summary_code_bot == true);
    var B = (_v('gbmscore_oms_accept') >= 0.9);
    var C = (_v('can_bankauth_cust_error_count_30d') >= 5);
    var D = (_v('order_amount') >= 100);
    return A && B && C && D;
    
        Explanation:

        Based on our meta rules.
        
        - Based on ThreatMetrix data, this order is likely to be one placed by bot.
        - Number of orders placed by the customer got denied by bank authorization failed more than 5 times in the past 30 days.
        - The order amount is more than 100 Candian Dollars.
            
    Output requirement:
    - Provide your output with a brief summary about the explanation of the meta rule.
    - Your tone should be FORMAL and understandable to business executives.
    - Your answers shall help business executives in identifying the reason for denail/acceptance of thier orders

    <Output format>
    Provide summary in bried two or three. sentences
    - Use bullet points for reasons for chargeback denail
    - all the bullet points should be unique and not repeated
    - Do not expose values of the variables, only mention the thresholds and validation checks
    </Output format>

    Once you have the final answer, say:
        Final Answer: <answer>

        Question: {input}
        """
    }
```

```python
from langchain.agents import initialize_agent, Tool

class MetaRuleExplanationAgent:
    def __init__(self) -> None:
        self.llm = llm

        self.variable_dict_tool = Tool(
            name="VariableDict",
            func=get_variable_descriptions,
            description=(
                "Retrieve the list of descriptions of a given list of variables. Input shall be a list of variables"             
            ),
            input_variables=["variables_list"],
        )
    
    def run(self, rule: str, verbose=True):
            tools = [ self.variable_dict_tool]
            self.agent = initialize_agent(
                    tools, 
                    self.llm, 
                    agent="zero-shot-react-description", 
                    verbose=verbose,
                    handle_parsing_errors=True,
                    agent_kwargs={
                        "prefix": prompts.get("PREFIX"),
                        "suffix": prompts.get("SUFFIX")    
                        },
            )
            
            return self.agent.run(
                f"Explain the meta rule {rule} \n"
    )
    
```
