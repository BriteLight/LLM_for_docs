```python
import os
import ssl
import httpx

from langchain_community.llms import HuggingFaceTextGenInference 

API_KEY = "YOUR_API_KEY"

WMT_CA_PATH = "ca-bundle.crt"

os.environ['SSL_CERT_FILE'] = WMT_CA_PATH
os.environ['REQUESTS_CA_BUNDLE'] = WMT_CA_PATH

headers =  { 
            "X-Api-Key": API_KEY,
            'Content-Type': 'application/json'
            }
context = ssl.create_default_context()
context.load_verify_locations(WMT_CA_PATH)
client =  httpx.Client(verify=context, headers=headers)



LLAMA_LLM_MODEL_NAME = "llama3-70b-instruct"
LLAMA_LLM_API_VERSION  = "00ffbd"


llm = HuggingFaceTextGenInference(
            inference_server_url=f"https://llmgateway.<priv-dom>.com/llmgateway/model/{LLAMA_LLM_MODEL_NAME}",
            max_new_tokens=2023,
            server_kwargs={
                "headers": headers
                })


llm("Hi")


```
